## Creating Box plot &Histogram & vilon plot on Continuous variables

```python
df.head()
```

```python
#Histogram on km driven
plt.hist(df['KM_Driven'])
```

```python
#Distribution plot on KM driven
sns.distplot(df['KM_Driven'], kde = True, bins = [20000,40000,60000,80000,100000])
plt.title('No of km driven')
plt.show()
```

```python
#box plot on km driven
df.boxplot(column = 'KM_Driven')
```

```python
#interquantile range
Q1 = df['KM_Driven'].quantile(0.25)
Q3 = df['KM_Driven'].quantile(0.75)
IQR = Q3-Q1
IQR
```

```python
#Violin plot on KM driven
sns.violinplot(df['KM_Driven'])
```
